## Q3: 何謂內部報酬率法（Internal Rate of Return, IRR）?決策準則為何?請說明使用IRR 評估投資計畫有那些缺點或限制(請至少寫出3點)

1. 內部報酬率法(IRR)是估算NPV=0時，報酬率為多少，以此報酬率和折現值做比較，來評估是不是要進行投資 <br />
2. 缺點與限制 <br />
   a. 投資與借貸的差異, 如果是借貸的話，NPV會有差，所仍需要計算NPV去確認計劃是否可行 <br />
   b. 多個IRR, 有可能有2個多重解 <br />
   c. 不存在IRR, 如果以一個計畫根本不存在NPV=0的話，IRR無法計算  <br />
   

## Q4:何謂獲利能力指數（Profitability Index, PI）?決策準則為何?獲利能力指數評估投資計畫的缺點為何?

1. 當有預算限制需要決定投資優先順序的時候，單用NPV無法判定，PI可以在有預算限制下做出最佳決策，PI = 第一期開始之現金流量總現值/初期投資<br />
2. 缺點 : 當互斥計畫下，會忽略期初投資規模

## Q5: 資本預算決策中現金流量的估算是以現金為基礎，為何需考慮折舊費用?

因為折舊稅盾關係，折舊會影響到公司應繳稅額，當如果公司稅是20%,折舊1元，應繳稅金降低0.2元，現金流增加0.2元
